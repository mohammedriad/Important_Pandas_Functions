# group by Function 
####  تساعدنه هاي الفكشن نجمع الداتا مالتنه حسب عمود معين او فيتشر معين بالداتا و نسوي عليه عمليات معينه (مثل نحسب المعدل او نحسب العدد)
#### او نطبق فكشن معينه ع الداتا سيت 

```python
import pandas as pd 
```

```python
data   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas","Hadoop","Spark","Python","NA"],
    'Fee' :[22000,25000,23000,24000,26000,25000,25000,22000,1500],
    'Duration':['30days','50days','55days','40days','60days','35days','30days','50days','40days'],
    'Discount':[1000,2300,1000,1200,2500,None,1400,1600,0]
          })
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby(["Courses"]).mean()
```

```python
df.groupby(["Courses"]).sum()
```

```python
# اذا نريد نطبقها على اكثر من عمود نكدر نسوي هيج 
df.groupby(["Courses","Duration"]).sum()
```

```python
# نلاحظ بالامثله الفوك ماكو اندكس للصفوف ف اذا نريد نظهره نستعمل 
df.groupby(["Courses","Duration"]).sum().reset_index()
```

```python
# اذا كان عدنه الكيز الي دنقسم او نجمع الداتا ع اساسهن مفقودات او نل و نريد نعرضهن نستعمل هذا الباراميتر 
df.groupby(["Courses"],dropna=False).sum()
```

```python
# اذا تريد النتيجه ما تترتب و هيج حيكون الاداء اسرع 
df.groupby(["Courses"],sort=False).sum()
```

```python
# بس نكدر نرتب الداتا بشكل تنازلي يعني من الجبير للصغير ب استعمال 
# رتبتهن على حسب نفس الداتا الفوك
sorted_data = df.sort_values(['Courses'], ascending=False)
sorted_data
```

```python
# و ايضا نكدر نستعمل هاي الفكشن حته نطبق فكشن معينه ع الداتا سيت مالتنه 
# تنطي نفس الناتج 
sorted_df = df.apply(lambda x : x.sort_values(ascending=False))
sorted_df
```

```python
# اما اذا تريد تجمع مثلا عمودين بس من الداتا و كل عمود تسوي عليه عمليه معينه ف نستعمل
x=df.groupby(['Courses','Duration']).aggregate(['count','max'])
```

```python
# و اذا كان شكلهن مو حلو نكدر نعدله من خلال هذا الكود البسيط 
x.columns = ["_".join(col).strip() for col in x.columns.values]
x
```

```python
# و ايضا نكدر نجمع الداتا حسب عمود معين و نسوي عمليه ع الداتا مثل نحسب او نشوف المعدل ع اساس عمود ثاني 
# مثلا هنا جمعت الداتا على حسب عمود الكورسات و حسبت مجموع الاستقطاعات لكل كورس بسهوله 
df.groupby('Courses')['Discount'].sum()
```

```python
# جمعت الداتا حسب عمود الكورسات و اختاريت عمودين و طبقت عليهم ميثود الفيرست الي ترجعلك اول قيمه لكل كورس 
df.groupby('Courses')[['Fee','Discount']].first()
```

```python
# هنا طبقت اكثر من عمليه مختلفه لكل عمود مختلف
a = df.groupby('Courses')
a_agg =a.agg({
    'Discount':'sum',
    'Fee':'min',
    'Duration':'count'
    })
a_agg
```

```python
# نكدر نستعمل الفكشن فلتر ايضا هي مثل ابلي 
# و لكن لازم الفكشن الي تخلي بيها لازم يكون بيها شرط معين حته تتنفذ

# هاي الفكشن
def filter_fuc(datasets):
    datasets['Discount'].mean()>5
```

```python
df.groupby('Courses').filter(filter_fuc)
```
