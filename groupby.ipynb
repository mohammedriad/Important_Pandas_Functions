# Groupby_Function

```python
# import Necessary libraries
import pandas as pd
import numpy as np 
```

##### read the dataFrame by `read_csv`

```python
df = pd.read_csv("heart.csv")
df
```

##### `Groupby_Function` its a pandas function that help you to group your data by the unique data from one columns

```python
df.groupby("exang").mean()

# عمليه جمع البيانات تكون من خلال انه ياخذ قيم العناصر من غير تكرارها(الموجوده في العمود) و يحسب قيمها 
# و يقسم بقيه الداتا فريم حسب قيم اليونيك مال العمود المختار 
# و ايضا مثل ما تلاحظ هنا نكدر نسوي ع البيانات عمليات معينه 
```

##### and we can pick some of the data that we want to group it with a sertain column  
##### `this is one of many strategies of how to do it `

```python
df.groupby("exang")[["age","chol"]].first()

# ايضا نستطيع اختيار جزء من الداتا لجمعها مع عمود معين و اجراء بعض العمليات عليها 
# first finction :  وظيفتها تعطيك اول قيمه من كل عمود 
```

#### `agg Function` it help us to make more than one opeartion and on diffrent columns in the same time 
###### so here in this example we group our data by `cp_column` and stor it in `g_df`  and  by using `agg_function`
###### we select a specific coulumn and do a specific operation on it 

```python
g_df = df.groupby('cp')

a = g_df.agg({
    "sex":["mean","min"],
    "fbs": "first",
    "chol":"size"
})
a
#استعملت الفكشن و خليت بيها دكشنري حيث الكي يمثل اسم العمود و الفاليو تمثل العمليه الي اريد اسويه ع العمود 
# بس اكو مشكله صغيره الي هيه اسماء الاعمده شكلهن مو حجلو ف نكدر نعدلها من خلاال 
```

#### we notic that there is some mistake with the names of the columns so we use this little trick  to solve this problem 

```python
a.columns = ["__".join(col).strip() for col in a.columns.values]
a

# شكلهن ترتب 
```
